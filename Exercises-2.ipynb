### What are the ten most common movie names of all time?

```python
top_10_common_titles = titles.title.value_counts().head(10)
```

```python
print("The ten most common movie names of all time:")
print(top_10_common_titles )
```

### Which three years of the 1930s saw the most films released?

```python
top_3_years_1930s = titles[titles.year //10 == 193].year.value_counts().head(3)
```

```python
print("The three years of the 1930s with the most films released:")
print(top_3_years_1930s)
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
titles['decade'] = ((titles.year // 10 )*10)
titles.decade.value_counts().sort_index().plot(kind='bar')
plt.title('Number of films that have been released each decade over the history of cinema')
plt.xlabel('Decade')
plt.ylabel('Count')
plt.show()
```

### Plot the number of "Hamlet" films made each decade.

```python
titles[titles.title == 'Hamlet'].decade.value_counts().sort_index().plot(kind='bar')
plt.title('Number of "Hamlet" Flims in Each Decade')
plt.xlabel('Decade')
plt.ylabel('Count')
```

### Plot the number of "Rustler" characters in each decade of the history of film.

```python
cast['decade'] = ((cast.year// 10)*10)
cast[cast.character == 'Rustler'].decade.value_counts().sort_index().plot(kind='line')
plt.title('Number of "Rustler" Characters in Each Decade')
plt.xlabel('Decade')
plt.ylabel('Count')
```

### Plot the number of "Hamlet" characters each decade.

```python
cast[cast.character == 'Hamlet'].decade.value_counts().sort_index().plot(kind='barh')
plt.title('Number of "Hamlet" Characters in Each Decade')
plt.xlabel('Decade')
plt.ylabel('Count')
```

### Plot the n-value of the roles that Sidney Poitier has acted over the years.

```python
c=cast[(cast.name == 'Sidney Poitier')].n.value_counts().sort_index()
c.plot(kind='line', title='Distribution of "n"-Values for Sidney Poitier\'s Roles')
plt.xlabel('"n"-Value')
plt.ylabel('Number of Roles')
plt.grid(axis='y')
plt.show()
```
